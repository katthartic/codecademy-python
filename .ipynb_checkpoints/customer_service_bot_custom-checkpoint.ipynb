```python
#Global Variables
response_number_text = "\nPlease enter the number corresponding to your choice: "
error_text = "\nSorry, we didn't understand your selection.\n\n"
```

```python
def cs_service_bot():
    #Main function to initiate custom service bot.
    #Inputs commented below...
    print("Hello! Welcome to the DNS Cable Company's Service Portal. Are you a new or existing customer?\n [1] New Customer\n [2] Existing Customer")
    response = input(response_number_text)
    if response == "1": #New Customer
        new_customer()
    elif response == "2": #Existing Customer
        existing_customer()
    else:
        print(error_text)
        cs_service_bot()
```

```python
def existing_customer():
    #Function for existing customer service options
    #Inputs commented below...
    print("\nWhat kind of support do you need?\n[1] Television Support\n[2] Internet Support\n[3] Speak to a support representative")
    response = input(response_number_text)
    if response == '1': #Television Support
        television_support()
    elif response == '2': #Internet Support
        internet_support()
    elif response == '3': #Speak to a support representative
        live_rep("support")
    else:
        print(error_text)
        existing_customer()
```

```python
def new_customer():
    #Function for new customer service options
    #Inputs commented below...
    print("\nWe're excited to have you join the DNS family, how can we help you?\n[1] Sign up for service.\n[2] Schedule a home visit.\n[3] Speak to a sales representative.")
    response = input(response_number_text)
    if response == '1': #Sign up for service
        sign_up()
    elif response == '2': #Schedule a home visit
        home_visit()
    elif response == '3': #Speak to a sales representative
        live_rep("sales")
    else:
        print(error_text)
        new_customer()
```

```python
def television_support():
    #Function for television support responses
    #Inputs commented below...
    print("\nWhat is the nature of your problem?\n[1] I can't access certain channels.\n[2] My picture is blurry.\n[3] I keep losing service.\n[4] Other issues.")
    response = input(response_number_text)
    if response == '1': #I can't access certain channels
        print("\nPlease check the channel lists at DefinitelyNotSinister.com. If the channel you cannot access is there, please contact a live representative.")
        did_that_help()
    elif response == '2': #My picture is blurry
        print("\nTry adjusting the antenna above your television set.")
        did_that_help()
    elif response == '3': #I keep losing service
        print("\nIs it raining outside? If so, wait until it is not raining and then try again.")
        did_that_help()
    elif response == '4': #Other issues
        live_rep("support")   
    else:
        print(error_text)
        television_support()
```

```python
def internet_support():
    #Function for internet support responses
    #Inputs commented below...
    print("\nWhat is the nature of your problem?\n[1] I can't connect to the internet.\n[2] My connection is very slow.\n[3] I can't access certain sites.\n[4] Other issues.")
    response = input(response_number_text)
    if response == '1': #I can't connect to the internet
        print("\nUnplug your router, then plug it back in, then give it a good whack, like the Fonz.")
        did_that_help()
    elif response == '2': #My connection is very slow
        print("\nMake sure that all cell phones and other peoples computers are not connected to the internet, so that you can have all the bandwidth.")
        did_that_help()
    elif response == '3': #I can't access certain sites
        print("\nMove to a different region or install a VPN. Some areas block certain sites.")
        did_that_help()
    elif response == '4': #Other issues
        live_rep("support")   
    else:
        print(error_text)
        internet_support()
```

```python
def did_that_help():
    #Function to assess support service provide
    #Inputs commented below...
    print("\nDid the solution work?\n[1] Yes\n[2] No")
    response = input(response_number_text)
    if response == '1': #Yes
        print("\nThank you! Have a great day!")
    elif response == '2': #No
        print("\nI'm sorry to hear that. Would you rather talk to a live representative or schedule a home visit?\n[1] Live Representative\n[2] Home Visit")
        response = input("\nPlease enter the number corresponding to your choice: ")
        if response == '1': #Live Representative
            live_rep("support")
        elif response == '2': #Home Visit
            home_visit("support")
        else:
            print("\nSorry, we didn't understand your selection. Goodbye!\n\n")
    else:
        print(error_text)
        did_that_help()
```

```python
def sign_up():
    #Function for new service sign-up
    #Inputs commented below...
    print("\nGreat choice, friend! We're excited to have you join the DNS family! Please select the package you are interested in signing up for.\n[1] Bundle Deal (Internet + Cable)\n[2] Internet\n[3] Cable")
    response = input(response_number_text)
    if response == '1': #Bundle Deal
        print("\nYou've selected the Bundle Package! Please schedule a home visit and our technician will come and set up your new service.")
        home_visit("new install")
    elif response == '2': #Internet
        print("\nYou've selected the Internet Only Package! Please schedule a home visit and our technician will come and set up your new service.")
        home_visit("new install")
    elif response == '3': #Cable
        print("\nYou've selected the Cable Only Package! Please schedule a home visit and our technician will come and set up your new service.")
        home_visit("new install")  
    else:
        print(error_text)
        sign_up()
```

```python
def home_visit(purpose = "none"):
    #Function to schedule a home visit
    #Inputs commented below...
    if purpose == "none":
        print("\nWhat is the purpose of this home visit?\n[1] New service installation.\n[2] Existing service repair.\n[3] Location scouting for unserviced region.")
        response = input(response_number_text)
        if response == '1': #New service installation
            home_visit("new install")
        elif response == '2': #Existing service repair
            home_visit("support")
        elif response == '3': #Location scouting for unserviced region
            home_visit("scout")
        else:
            print("\nSorry, we didn't understand your selection.\n\n")
            home_visit()
    else:
        if purpose == "new install":
            service_type = "install your new service.\n"
        elif purpose == "support":
            service_type = "repair your existing service.\n"
        elif purpose == "scout":
            service_type = "scout the unserviced area.\n"
        visit_date = input("\nPlease enter a date below when you are available for a technician to come to your home and " + service_type)
        print("\nWonderful! A technical will come visit you on " + visit_date + ". Please be available between the hours of 1:00 am and 11:00 pm.")
        return visit_date
```

```python
def live_rep(purpose):
    #Function to speak with rep. Accepts only "sales" or "service" via other functions.
    service_message = "/nPlease hold while we connect you with a live sales representative. The wait time will be between two minutes and six hours. We thank you for your patience."
    if purpose == "sales":
        print(service_message)
    elif purpose == "service":
        print(service_message)
    else:
        print(service_message)
```
